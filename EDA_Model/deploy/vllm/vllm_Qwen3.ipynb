## Qwen3-8B-IC 模型部署

多卡 openai格式

 http://localhost:8000

think 模式

使用环境变量 LLM_AGENT_MODEL_DIR 设置模型目录

CUDA_VISIBLE_DEVICES=6,7 \
python -m vllm.entrypoints.openai.api_server \
  --served-model-name Qwen3-8B-IC \
  --model ${LLM_AGENT_MODEL_DIR:-/data/llm_agent/models}/Qwen3-8B-IC \
  --dtype auto \
  --port 8000 \
  --tensor-parallel-size 2 \
  --pipeline-parallel-size 1 \
  --max-model-len 16384 \
  --gpu_memory_utilization 0.85 \
  --max-num-seqs 128 \
  --trust-remote-code \
  --enforce-eager

no think 模式

使用环境变量 LLM_AGENT_MODEL_DIR 设置模型目录

CUDA_VISIBLE_DEVICES=6,7 \
python -m vllm.entrypoints.openai.api_server \
  --served-model-name Qwen3-8B-IC \
  --model ${LLM_AGENT_MODEL_DIR:-/data/llm_agent/models}/Qwen3-8B-IC \
  --dtype auto \
  --port 8000 \
  --tensor-parallel-size 2 \
  --max-model-len 16384 \
  --gpu_memory_utilization 0.85 \
  --max-num-seqs 128 \
  --trust-remote-code \
  --enforce-eager \
  --reasoning-parser qwen3 \
  --default-chat-template-kwargs '{"enable_thinking": false}'

单卡 request 格式

vllm serve \
  --served-model-name Qwen3-8B-IC \
  --model ${LLM_AGENT_MODEL_DIR:-/data/llm_agent/models}/Qwen3-8B-IC \
  --port 8000 \
  --max-model-len 16384 \
  --reasoning-parser qwen3 \
  --default-chat-template-kwargs '{"enable_thinking": false}' \
  --gpu_memory_utilization 0.8 \
  --enforce-eager
