```python
import json
from datasets import load_dataset,Dataset
import os

def load_json_to_dataset(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)  # 加载为Python列表
    
    # 转换为Dataset对象
    return Dataset.from_list(data)

BASE_DATA_DIR = os.getenv("LLM_AGENT_DATA_DIR", "/data/llm_agent/data")
medical_dataset = load_json_to_dataset(os.path.join(BASE_DATA_DIR, "medical/combined_medical_data.json"))
medical_dataset[0]
```
