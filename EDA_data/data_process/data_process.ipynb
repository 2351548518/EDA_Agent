## 2 读取 数据集

```python
from datasets import load_dataset
from IPython.display import display, JSON
import os

# 基础数据目录，可通过环境变量覆盖
BASE_DATA_DIR = os.getenv("LLM_AGENT_DATA_DIR", "/data/llm_agent/data")

# 加载数据集
IC_dataset = load_dataset("SeekBCD/IC-Textbook_Dataset", cache_dir=os.path.join(BASE_DATA_DIR, "IC_Textbook_Dataset"),split="train")
IC_dataset
display(JSON(IC_dataset[0]))
```

```python
from datasets import Dataset
import os

BASE_DATA_DIR = os.getenv("LLM_AGENT_DATA_DIR", "/data/llm_agent/data")

Verilog_dataset = Dataset.from_file(os.path.join(BASE_DATA_DIR, "MG_Verilog/merged_dataset/data-00000-of-00001.arrow"))
Verilog_dataset
display(JSON(Verilog_dataset[0]))
```
